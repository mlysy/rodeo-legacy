```python
import numpy as np
import BayesODE as pos # probabilitic ODE solver
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f(x,t):
    return  3*(t+1/4) - x/(t+1/4)

a = 0
b = 4.75
x0 = 10
N = 100
tseq = np.linspace(a, b, N)
w = gamma = (b-a)/(N-1)
alpha = N/10

# package functions
Sigma_vv, Sigma_xx, Sigma_xv = pos.cov_prior(tseq, 'exp2', gamma, alpha)
xt1 = pos.ode_bayes(f, tseq, x0, Sigma_vv, Sigma_xx, Sigma_xv)

Sigma_vv, Sigma_xx, Sigma_xv = pos.cov_prior(tseq, 'rect', gamma, alpha)
xt2 = pos.ode_bayes(f, tseq, x0, Sigma_vv, Sigma_xx, Sigma_xv)

plt.plot(tseq, xt1, label = 'exp2')
plt.plot(tseq, xt2, label = 'rect')
plt.legend(loc='upper left')


x = odeint(f,x0,tseq)
plt.scatter(tseq, x, facecolor = 'none', edgecolor = 'b', label='analytical')
plt.legend(loc='upper left')
plt.show()
```
